```r
sim_test <- function(i) {
    n <- 100000
    c1 <- rnorm(n = n, mean = 3, sd = 1)
    p <- 0.5
    x <- rbinom(n = n, size = 2, prob = p)
    var_x <- 2 * p * (1-p)
    e <- rnorm(n = n, mean = 0, sd = 100)
    y <- 3 * x + e
    fit <- lm(y ~ x)
    beta <- coef(summary(fit))["x", "Estimate"]
    se <- coef(summary(fit))["x", "Std. Error"]
    tstat <- coef(summary(fit))["x", "t value"]
    pval <- coef(summary(fit))["x", "Pr(>|t|)"]
    
    # p = 3 because intercept, x, c1
    2 * pt(abs(beta/se), df = n - 2, lower.tail = FALSE) == pval
    resid_e_est <- sqrt(se^2 * n * var_x)
    resid_e_est
    x_sim <- rbinom(n = n, size = 2, prob = p)
    e_sim <- rnorm(n = n, mean = 0, sd = resid_e_est)
    y_sim <- 2 + beta * x_sim + e_sim
    fit_sim <- lm(y_sim ~ x_sim)
    beta_sim <- coef(summary(fit_sim))["x_sim", "Estimate"]
    se_sim <- coef(summary(fit_sim))["x_sim", "Std. Error"]
    tstat_sim <- coef(summary(fit_sim))["x_sim", "t value"]
    pval_sim <- coef(summary(fit_sim))["x_sim", "Pr(>|t|)"]
    ret <- data.table(pval = pval, pval_sim = pval_sim, tstat = tstat, tstat_sim = tstat_sim, beta = beta, beta_sim = beta_sim, se = se, se_sim = se_sim)
}
```

```r
profvis({tmp <- lapply(1:1000, sim_test) |> rbindlist()})
```

```r
ggplot(tmp, aes(x = -log10(pval), y = -log10(pval_sim))) +
    geom_point() +
    geom_abline(slope = 1, intercept = 0, color = "red")
```

```r
df <- fread(f)
```

```r
df_signif <- df[pval < 5e-8]
nrow(df_signif)
```

```r
nrows <- 100
pb <- txtProgressBar(min = 0, max = nrows, style = 3)
set.seed(1)
make_sim <- function(i, df){
    setTxtProgressBar(pb, i)
    the_snp <- df[i]
    variant <- the_snp[["variant"]]
    n <- the_snp[["n_complete_samples"]]
    maf <- the_snp[["minor_AF"]]
    beta <- the_snp[["beta"]]
    beta_se <- the_snp[["se"]]
    # var(x) computed from the binomial definition as n * p * (1 - p) with n = 2 and p = maf
    var_x <- 2 * maf * (1 - maf)
    # Since: beta_se = resid_se / sqrt(n * var(x))
    # I can derive the standard deviation of the residuals:
    # beta_se^2 = var(resid) / (n * var(x))
    # var(resid) = beta_se^2 * n * var(x)
    # resid_sd = sqrt(beta_se^2 * n * var(x))
    resid_sd <- sqrt(beta_se^2 * n * var_x)
    e <- rnorm(n = n, mean = 0, sd = resid_sd)
    x <- rbinom(n = n, size = 2, prob = maf)
    y <- beta * x + e
    coef <- lm(y ~ x) |> summary() |> coef()
    pval <- coef["x", "Pr(>|t|)"]
    tstat <- coef["x", "t value"]
    ret <- data.table(variant = variant, tstat_sim = tstat, pval_sim = pval)
    return(ret)
}
sim_res <- lapply(1:nrows, make_sim, df = df_signif) |> rbindlist()
close(pb)
```
